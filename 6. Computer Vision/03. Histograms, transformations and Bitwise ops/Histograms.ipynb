# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Histograms

```python
def imshow(img):
    plt.figure(figsize = (20,15));
    plt.imshow(img[:,:,::-1]);
```

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc
img = cv2.imread('img/blue-flowers.jpg', 1)
his_b = cv2.calcHist([img],[0],None,[255],[0,255])
his_g = cv2.calcHist([img],[1],None,[255],[0,255])
his_r = cv2.calcHist([img],[2],None,[255],[0,255])
imshow(img)
```

```python
plt.plot(his_b, color='b')
```

```python
plt.plot(his_g, color='g')
```

```python
plt.plot(his_r, color='r')
```

[Hint](# "cv2.calcHist") 


```python
# Now try to plot all 3 channels at the same time, does the histogram give you an idea of what the image looks like?
colors = ['b', 'g', 'r']
for i, color in enumerate(colors):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist, color=color)
    plt.xlim(0,256)
plt.show()
```
