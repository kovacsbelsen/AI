### Lets fix the colors

```python
# Does your image look ok? are the colors the right ones?, if not, lets fix it!
# Remember that openCV will load images using BGR rather than RGB
# Change the image to RGB and diplay it again with plt
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

[Hint](# "use cvtColor") 

### What is the shape of the image?

### Lets make it grayscale

```python
# Load the the image in grayscale, what are the 2 ways you can do that?
grey_img = cv2.imread('image.jpg',0)  # flag 0 = grey
cv2.imshow('image',grey_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Now load an image in color and then transform it to grayscale
greyscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('image',greyscale_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### What is the shape now?

### We need to indicate the colormap

```python
# Display the grayscale image with matplotlib, make sure to include the colormap so it really is grayscale
plt.imshow(grey_img, cmap='gray');
```

### Cropping an image

```python
# Lets now crop a Region of Interest of an image, load and image and crop different parts of it
crop_img = img[90:550, 120:880]
cv2.imshow("Image", crop_img)
cv2.waitKey(0)
```

```python
# Save to disk the cropped areas
filename = 'croppedImage.jpg'
cv2.imwrite(filename, crop_img)
```

### To edit an image making a copy of it is always a good idea!

```python
# Load an image and then make a copy of it
img_copy = crop_img.copy()
img_copy = cv2.cvtColor(img_copy,cv2.COLOR_BGR2RGB)
plt.imshow(img_copy)
```

[Hint](# "use the numpy function ") 

## Annotating images

### Rectangles

```python
# Draw green rectangle on top of a region of interest in your image
green = (0,255,0)
cv2.rectangle(img_copy,(15,200),(320,450), green, 3)
plt.imshow(img_copy)
```

```python
red = (255, 0, 0)
cv2.rectangle(img_copy,(450,5),(760,450), red, 3)
plt.imshow(img_copy)
```

[Hint](# "You will need to use the cv2 function rectangle(img, pt1, pt2, color, thickness) -> img") 

### Adding text to images

```python
# Make a copy of your image and then write some text on top of the image
cv2.putText(img_copy, 'Doggy', (100,170), cv2.FONT_ITALIC, 2, green, cv2.LINE_4)
plt.imshow(img_copy)
```

```python
cv2.putText(img_copy, 'Person', (550,60), cv2.FONT_ITALIC, 2, red, cv2.LINE_4)
plt.imshow(img_copy)
```

[Hint](# "You will need to use cv2 putText function -> putText(img, text, coordinates, fontFace, fontScale, color, thickness, lineType). For fonts check the enum cv2.FONT_HERSHEY_")

```python
# Load as many images as you like and start annotating objects in the image, 
# Before annotating the images make a copy of the image so we keep the original without changes
# Use the format proposed in the example above

```

```python
# Display the images that you have annotated above, both versions, the original one and the annotated one
```

```python
img2 = cv2.imread('street.jpg',1)
img2_copy = img2.copy()
img2_copy = cv2.cvtColor(img2_copy,cv2.COLOR_BGR2RGB)

black = (0,0,0)
green = (0,255,0)
red = (255, 0, 0)
cv2.rectangle(img2_copy,(200,230),(400,640), green, 3)
cv2.rectangle(img2_copy,(200,200),(400,230), green, -1)
cv2.putText(img2_copy, 'Cyclist', (210,220), cv2.FONT_ITALIC, 1, black, cv2.LINE_4)

cv2.rectangle(img2_copy,(500,240),(600,400), red, 3)
cv2.rectangle(img2_copy,(500,240),(600,200), red, -1)
cv2.putText(img2_copy, 'Walker', (500,230), cv2.FONT_ITALIC, 1, black, cv2.LINE_4) # walker = pedestrian just shorter

cv2.rectangle(img2_copy,(640,260),(720,440), red, 3)
cv2.rectangle(img2_copy,(640,260),(720,220), red, -1)
cv2.putText(img2_copy, 'Walker', (630,250), cv2.FONT_ITALIC, 1, black, cv2.LINE_4)


cv2.rectangle(img2_copy,(725,260),(800,450), red, 3)
cv2.rectangle(img2_copy,(725,260),(800,220), red, -1)
cv2.putText(img2_copy, 'Walker', (730,250), cv2.FONT_ITALIC, 1, black, cv2.LINE_4)




plt.imshow(img2_copy)

img2_copy = cv2.cvtColor(img2_copy,cv2.COLOR_RGB2BGR)

filename = 'annotated.jpg'
cv2.imwrite(filename, img2_copy)
```
