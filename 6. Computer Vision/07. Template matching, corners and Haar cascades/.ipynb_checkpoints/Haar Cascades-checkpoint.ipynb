```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def imshow(img, enlarge = True, color = True):
    if enlarge:
        plt.figure(figsize = (30,20));
    if not color:
        plt.imshow(img, cmap='gray');
    else:
        plt.imshow(img[:,:,::-1]);
```

```python
image = cv2.imread('img/people2.jpg');
image2 = image.copy()
imshow(image)
```

```python
# Given the image above, lets detect faces on the image using haar cascades
# First create a CascadeClassifier using the frontalFace classifier -> Haarcascades/haarcascade_frontalface_default.xml


face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')

# Convert the image to grayscale

g_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


# Using the classifiers detect all faces on the image

faces = face_classifier.detectMultiScale(image, 1.3, 5)

# If no faces detected print("No faces found")

if faces is():
    print('no faces found')

# Otherwise draw a rectangle on each face that has been detected

for (x,y,w,h) in faces:
    face = g_img[y:y+h, x:x+w]
    cv2.rectangle(image, (x,y), (x+w,y+h), (255,255,0),2)
    imshow(image,False)
```

```python
# Repeat the process above with your own images
image1 = cv2.imread('img/selfie.jpg');

face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
faces = face_classifier.detectMultiScale(image1, 1.3, 5)

g_img1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)

if faces is():
    print('no faces found')
    
for (x,y,w,h) in faces:
    face = g_img1[y:y+h, x:x+w]
    cv2.rectangle(image1, (x,y), (x+w,y+h), (255,255,0),5)
    
imshow(image1,False)

```

```python
# Now do the same but this time also detect eyes in the faces using -> Haarcascades/haarcascade_eye.xml

face_cascade = cv2.CascadeClassifier("Haarcascades/haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')


eyes = eye_classifier.detectMultiScale(image2)


img = cv2.imread("img/friends.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray, 1.3, 5)


for (x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
