```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

data = pd.read_csv('company_sales_data.csv')
#print(data)

profit = data['total_profit'].tolist()
month = data['month_number'].tolist()


plt.plot(month, profit, label = 'Profit of last 12 months')

plt.xlabel('Month nmber')
plt.ylabel('Profit in dollar')
plt.legend(loc='lower right')
plt.title('Company profit per month')
plt.xticks(month)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```
